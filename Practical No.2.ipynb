

*  Byte objects vs string in python


```python
# Encoding

var1 = 'hii'
byteform = b'hii'
var2 = var1.encode('ASCII')

if(var2 == byteform):
   print("Encoding done")        
else: 
   print('Encoding not done yet')   
```

```python
#Decoding

var1 = 'hii'
byteform = b'hii'
var2 = byteform.decode('ASCII')

if(var2 == var1):
   print("Decoding done")        
else: 
   print('Decoding not done yet') 
```
